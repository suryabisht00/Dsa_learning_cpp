## ğŸ§© 2. `std::vector`

### ğŸ”¹ Vector-Specific Functions
| Function         | Description                    |
|------------------|--------------------------------|
| `push_back(x)`   | Add element at end             |
| `pop_back()`     | Remove last element            |
| `at(i)`          | Element at index `i`           |
| `front()` / `back()` | First / last element       |

### ğŸ”¹ Example

## ğŸ§© 3. `std::deque`

### ğŸ”¹ Deque-Specific Functions
| Function         | Description                       |
|------------------|-----------------------------------|
| `push_front(x)`  | Insert at beginning               |
| `push_back(x)`   | Insert at end                     |
| `pop_front()`    | Remove from front                 |
| `pop_back()`     | Remove from back                  |
| `front()` / `back()` | Access front/back             |

### ğŸ”¹ Example
```cpp
std::deque<int> d;
d.push_front(1);
d.push_back(2);
d.pop_back();
```

## ğŸ§© 4. `std::queue`

### ğŸ”¹ Queue-Specific Functions
| Function     | Description                          |
|--------------|--------------------------------------|
| `push(x)`    | Enqueue at back                      |
| `pop()`      | Dequeue from front                   |
| `front()`    | Get front element                    |
| `back()`     | Get last element                     |

### ğŸ”¹ Example
```cpp
std::queue<int> q;
q.push(1);
q.pop();
```

## ğŸ§© 5. `std::set` (Ordered)

### ğŸ”¹ Set-Specific Functions
| Function       | Description                     |
|----------------|---------------------------------|
| `insert(x)`    | Adds unique element             |
| `erase(x)`     | Removes element                 |
| `count(x)`     | Checks existence (0 or 1)       |
| `find(x)`      | Returns iterator or `end()`     |

### ğŸ”¹ Example

## ğŸ§© 6. `std::unordered_set`

### ğŸ”¹ Same as `set` but **unordered and faster (O(1)** on average)
- No order guarantee
- Based on hash table

### ğŸ”¹ Example
```cpp
std::unordered_set<int> us;
us.insert(5);
us.erase(5);
```

## ğŸ§© 7. `std::map` (Ordered)

### ğŸ”¹ Map-Specific Functions
| Function          | Description                     |
|-------------------|---------------------------------|
| `m[key] = value`  | Insert/update key-value         |
| `erase(key)`      | Removes by key                  |
| `count(key)`      | Check if key exists             |
| `find(key)`       | Returns iterator to key         |

### ğŸ”¹ Example
```cpp
std::map<int, std::string> m;
m[1] = "one";
if (m.count(1)) { /* exists */ }
```

## ğŸ§© 8. `std::unordered_map`

### ğŸ”¹ Unordered version of `map` (uses hash table)

- Fast lookups and insertions on average (O(1))

### ğŸ”¹ Example
```cpp
std::unordered_map<int, std::string> um;
um[2] = "two";
```
